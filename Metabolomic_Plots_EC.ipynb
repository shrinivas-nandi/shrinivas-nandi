```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# read in the updated files 
df = pd.read_csv (r'/Users/Shrini/Desktop/Australia_2023_Data/Metabolites/Acropora/Positive Metabolites/Acropora_Peak_Pos.csv')
df.replace(0, np.nan, inplace=True)
df = df[df.isnull().sum(axis=1) < 9]
Counts = df
######
df = df.set_index(['compound'])
Gene_Transposed = df.transpose()
Gene_Transposed = Gene_Transposed.reset_index()

####33
Gene_Transposed_TP = Gene_Transposed[Gene_Transposed['index'].str.contains('Hot')]
Gene_Transposed_TP = Gene_Transposed_TP.assign(Treatment="Hot")

Gene_Transposed_Amb = Gene_Transposed[Gene_Transposed['index'].str.contains('Amb')]
Gene_Transposed_Amb = Gene_Transposed_Amb.assign(Treatment = 'Amb')

frames = [Gene_Transposed_Amb, Gene_Transposed_TP]
result = pd.concat(frames)
result

#######
result_T0 = result[result['index'].str.contains('TP0')]
result_T0 = result_T0.assign(Time="0")

# for T2 

result_T1 = result[result['index'].str.contains('TP1')]
result_T1 = result_T1.assign(Time="1")

# for T3
result_T2 = result[result['index'].str.contains('TP2')]
result_T2 = result_T2.assign(Time="2")


frames = [result_T0, result_T1, result_T2]
df_wd = pd.concat(frames)
df_wd
```
