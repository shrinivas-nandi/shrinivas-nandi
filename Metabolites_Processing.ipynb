```python
# read in the updated files 
df = pd.read_csv (r'/Users/Shrini/Desktop/Australia_2023_Data/Metabolites/Acropora/Positive Metabolites/Acropora_Peak_Pos.csv')
df
```

```python
df.replace(0, np.nan, inplace=True)
```

```python
df = df[df.isnull().sum(axis=1) < 9]
Counts = df
Counts
```

```python
df = df.set_index(['compound'])
Gene_Transposed = df.transpose()
Gene_Transposed = Gene_Transposed.reset_index()
Gene_Transposed
```

```python
Gene_Transposed_TP = Gene_Transposed[Gene_Transposed['index'].str.contains('Hot')]
Gene_Transposed_TP = Gene_Transposed_TP.assign(Treatment="Hot")

Gene_Transposed_Amb = Gene_Transposed[Gene_Transposed['index'].str.contains('Amb')]
Gene_Transposed_Amb = Gene_Transposed_Amb.assign(Treatment = 'Amb')

frames = [Gene_Transposed_Amb, Gene_Transposed_TP]
result = pd.concat(frames)
result
```

```python
result_T0 = result[result['index'].str.contains('TP0')]
result_T0 = result_T0.assign(Time="0")

# for T2 

result_T1 = result[result['index'].str.contains('TP1')]
result_T1 = result_T1.assign(Time="1")

# for T3
result_T2 = result[result['index'].str.contains('TP2')]
result_T2 = result_T2.assign(Time="2")


frames = [result_T0, result_T1, result_T2]
df_wd = pd.concat(frames)
df_wd
```

```python
Name = "Glutathione disulfide" # RAN 

fig, ax = plt.subplots(figsize=(10, 8))

sns.stripplot(x = "Time",
              y = Name,
              hue = "Treatment",
              data = df_wd,
              ax = ax,
              dodge=True,
              s = 8,
              marker="D",
              palette="Set2",
              alpha = 0.7)

sns.boxplot(x = "Time",
            y = Name,
            data = df_wd,
            ax = ax,
            hue = "Treatment")

sns.despine(right = True)
ax.set( xlabel = "Experiment Time", ylabel = "Abundance")
plt.grid()
plt.title(Name,fontsize=12)
plt.suptitle('Glutathione disulfide',fontsize=16)


plt.show()
```

```python
Name = "Lysine" # RAN 

fig, ax = plt.subplots(figsize=(10, 8))

sns.stripplot(x = "Time",
              y = Name,
              hue = "Treatment",
              data = df_wd,
              ax = ax,
              dodge=True,
              s = 8,
              marker="D",
              palette="Set2",
              alpha = 0.7)

sns.boxplot(x = "Time",
            y = Name,
            data = df_wd,
            ax = ax,
            hue = "Treatment")

sns.despine(right = True)
ax.set( xlabel = "Experiment Time", ylabel = "Abundance")
plt.grid()
plt.title(Name,fontsize=12)
plt.suptitle('Lysine',fontsize=16)


plt.show()
```

```python
def save_combined_plots(dataframe):
    num_columns = len(dataframe.columns) - 1  # Subtract 2 for "Treatment" and "Time" columns

    for i, column in enumerate(dataframe.columns[:-1]):  # Exclude "Treatment" and "Time" columns
        fig, ax = plt.subplots(figsize=(10, 6))

        sns.stripplot(
            x="Time",
            y=column,
            hue="Treatment",
            data=dataframe,
            ax=ax,
            dodge=True,
            s=8,
            marker="D",
            palette="Set2",
            alpha=0.7
        )

        sns.boxplot(
            x="Time",
            y=column,
            data=dataframe,
            ax=ax,
            hue="Treatment",    
        )

        sns.despine(right=True)
        ax.set(xlabel="Experiment Time", ylabel="Abundance")
        plt.grid()
        plt.title(f"{column} - Stripplot and Boxplot", fontsize=12)
        plt.suptitle('Trial', fontsize=16)

        # Save the subplot as a separate file
        filename = f"{column}_combined_plot.png"
        plt.savefig(filename)
        plt.close()

# Call the function with your dataframe
save_path = "/Users/Shrini/Desktop/India Projects/PKS/Plots"

save_combined_plots(df_wd)
```
